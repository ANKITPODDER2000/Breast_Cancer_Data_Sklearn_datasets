```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import mglearn
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
X = data['data']
y = data['target']
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data=X,columns=data['feature_names'])
df['target'] = data['target']
```

<h1>Visualization</h1>

```python
sns.heatmap(df.corr())
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=101)
```

<h1>Logistic Regression</h1>

```python
from sklearn.linear_model import LogisticRegression

train_score = []
test_score = []
c = [0.0001,0.001,0.01,0.1,1,5,10,100,1000]
for i in c:
    lr = LogisticRegression(C=i).fit(X_train,y_train)
    print("Score for C = ",i)
    print("Test score for Train data : %.3f %%"%(lr.score(X_train,y_train) * 100))
    print("Test score for Test data : %.3f %%"%(lr.score(X_test,y_test) * 100))
    train_score.append(lr.score(X_train,y_train))
    test_score.append(lr.score(X_test,y_test))
    print("==============================")
```

```python
sns.set_style('darkgrid')
```

```python
plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
line, = plt.plot(c,train_score,'ro')
plt.setp(line,markersize=8)
plt.subplot(1,2,2)
line, = plt.plot(c,test_score,'bo')
plt.setp(line,markersize=8)
```

<h1>KNN Method</h1>

```python
from sklearn.neighbors import KNeighborsClassifier

train_score = []
test_score = []
c = list(range(1,11))
for i in c:
    knn = KNeighborsClassifier(n_neighbors=i).fit(X_train,y_train)
    print("Score for C = ",i)
    print("Test score for Train data : %.3f %%"%(knn.score(X_train,y_train) * 100))
    print("Test score for Test data : %.3f %%"%(knn.score(X_test,y_test) * 100))
    train_score.append(knn.score(X_train,y_train))
    test_score.append(knn.score(X_test,y_test))
    print("==============================")
```

```python
plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
line, = plt.plot(c,train_score,'ro')
plt.setp(line,markersize=8)
plt.subplot(1,2,2)
line, = plt.plot(c,test_score,'bo')
plt.setp(line,markersize=8)
```

<h1>Decision Tree</h1>

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import graphviz

train_score = []
test_score = []
c = list(range(1,11))
for i in c:
    dt = DecisionTreeClassifier(max_depth=i,random_state=0).fit(X_train,y_train)
    print("Score for C = ",i)
    print("Test score for Train data : %.3f %%"%(dt.score(X_train,y_train) * 100))
    print("Test score for Test data : %.3f %%"%(dt.score(X_test,y_test) * 100))
    train_score.append(dt.score(X_train,y_train))
    test_score.append(dt.score(X_test,y_test))
    print("==============================")
    
    
    
```

```python
plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
line, = plt.plot(c,train_score,'ro')
plt.setp(line,markersize=8)
plt.subplot(1,2,2)
line, = plt.plot(c,test_score,'bo')
plt.setp(line,markersize=8)
```

```python
export_graphviz(dt,out_file='mytree.dot',class_names=data['target_names'],feature_names=data['feature_names'],filled=True)
    
with open('mytree.dot') as file:
    content = file.read()
graphviz.Source(content)
```

<h1>Random Forest</h1>

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
n_estimator = [5,10,20,50,100]
random_state = [0,5,10,15,20]

train_score = []
test_score = []
label = []

for i in n_estimator:
    for j in random_state:
        model = RandomForestClassifier(n_estimators=i,random_state=j)
        model.fit(X_train,y_train)
        print("Model for n_estimator = %d and random_state = %d"%(i,j))
        train_score.append(model.score(X_train,y_train))
        test_score.append(model.score(X_test,y_test))
        print("Score for Train Data : %.3f %%"%(model.score(X_train,y_train) * 100))
        print("Score for Test Data : %.3f %%"%(model.score(X_test,y_test) * 100))
        print("----------------------------------------------------")
        label.append("n_estimators = %d & random_state = %d"%(i,j))
```
